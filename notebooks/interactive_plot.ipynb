```python
import sklearn.datasets
from sklearn import decomposition
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
digits = sklearn.datasets.load_digits()
data=digits['data']
labels=digits['target']
```

```python
view = TSNE(n_components=2).fit_transform(data)
```

```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
output_notebook()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(view, columns=['x','y'])
df['label'] = labels.astype('str')
df
```

```python
from bokeh.palettes import d3
from bokeh.models import CategoricalColorMapper
palette = d3['Category10'][10]
cmap = CategoricalColorMapper(
    factors=df['label'].unique(),
    palette=palette
)

fig_scat = figure()
source = ColumnDataSource(df)
fig_scat.scatter(
    x='x', y='y', alpha=0.5, 
    color={'field': 'label', 'transform': cmap}, 
    source=source
)
```

```python
from bokeh.layouts import row
images = data.reshape(data.shape[0],8,8)
fig_detail = figure(plot_width=300, plot_height=300, tools='')
fig_source = ColumnDataSource({'image':[]})
fig_detail.image('image', x=0, y=0, dw=10, dh=10, source=fig_source)
```

```python
from bokeh.models import CustomJS
from bokeh.models import HoverTool

code = """
var indices = cb_obj.indices;
console.log(indices);
"""
callback = CustomJS(args={'images': fig_source}, code=code)
hover = HoverTool(
    tooltips = None,
    callback = callback
)
fig_scat.add_tools(hover)
```

```python
layout = row(fig_scat,fig_detail)
show(layout)
```
